## MODELANDO: SMV

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
data_norm.head()
```

```python
x_data = data_norm.drop(["CUSTOMERID","CHECKINGSTATUS","CREDITHISTORY","LOANPURPOSE","EXISTINGSAVINGS","JOB","TELEPHONE","FOREIGNWORKER","RISK","RISK_n"], axis=1)
x_data.head()
```

```python
x_data1 = x_data.drop(["EMPLOYMENTDURATION","SEX","INSTALLMENTPLANS","HOUSING"], axis=1)
x_data1.head()
```

```python
x_data2=x_data1.drop(["OTHERSONLOAN","OWNSPROPERTY"], axis=1)
x_data2.head()
```

```python
y_data = data_norm["RISK_n"]
y_data.head()
```

```python
X_train, X_test, y_train,y_test = train_test_split(x_data2,y_data,test_size=0.3,random_state=42)
```

```python
svmModel = SVC(kernel="linear")
svmModel.fit(X_train,y_train)
```

```python
y_predict = svmModel.predict(X_test)
```

```python
cm = confusion_matrix(y_test, y_predict)
print(cm)
```
